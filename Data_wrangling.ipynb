# Camera location data

```python

df_camloc=pd.read_csv('Traffic_light_camera_locations_map.csv')

df_camloc
```

## Merging camera location with the offence data frame

```python

df_camloc1=df_camloc[['CAMERA TYPE','LOCATION_CODE','LATITUDE','LONGITUDE']]


# there are multiple locations for each speed camera '0001A','0001B', 
#create a column to indicate the camera with multiple locations(Mobile camera)
df_camloc1['mul_loc']=df_camloc1['LOCATION_CODE'].str.contains(r'\d\d\d\d\D')


# fixed camera location
df_singloc=df_camloc1[df_camloc1['mul_loc']==False]





#mobile camera with multiple locations
df_mulloc=df_camloc1[df_camloc1['mul_loc']==True]

#group the different type of the multiple location codes
#eg 0001A,0010A,0100A,1000A
df_mulloc['1']=df_mulloc['LOCATION_CODE'].str.contains('[0][0][0][1-9]')
df_mulloc['2']=df_mulloc['LOCATION_CODE'].str.contains('[0][0][1-9][0-9]')
df_mulloc['3']=df_mulloc['LOCATION_CODE'].str.contains('[0][1-9][0-9][0-9]')

dfA1=df_mulloc[df_mulloc['1']==True]
dfA1ls=list(dfA1['LOCATION_CODE'])

dfA2=df_mulloc[df_mulloc['2']==True]
dfA2ls=list(dfA2['LOCATION_CODE'])



dfA3=df_mulloc[df_mulloc['3']==True]
print(len(dfA3))
dfA3ls=list(dfA3['LOCATION_CODE'])



dfA4=df_mulloc[(df_mulloc['3']==False)&(df_mulloc['2']==False)&(df_mulloc['1']==False)]
print(dfA4)
dfA4ls=list(dfA4['LOCATION_CODE'])


#delete the prefix(0), and suffix from the location code
dfA1['LOCATION_CODE']=dfA1['LOCATION_CODE'].astype(str).str[3:4]
dfA2['LOCATION_CODE']=dfA2['LOCATION_CODE'].astype(str).str[2:4]
dfA3['LOCATION_CODE']=dfA3['LOCATION_CODE'].astype(str).str[1:4]
dfA4['LOCATION_CODE']=dfA4['LOCATION_CODE'].astype(str).str[0:4]

#pool all the data frame together
df_allloc=pd.concat([dfA1,dfA2,dfA3,dfA4,df_singloc])[['CAMERA TYPE','LOCATION_CODE','LATITUDE','LONGITUDE']]
```

```python
#checking whther there is any location with null value in the offence data frame
allcode=list(df_camloc1['LOCATION_CODE'])
allcode=[str(each) for each in allcode]
ct_na=0
for each in allcode:
    if each=='nan':
        ct_na+=1
print(ct_na)
```

```python
#rename the data frame for camera location data frame 
df_allloc = df_allloc.rename({'CAMERA TYPE': 'Camera_Type', 'LOCATION_CODE': 'Location_Code'}, axis='columns')

df_allloc=df_allloc.astype(str)

```

```python
#join the offence data frame and the camera location data frame together

df_offence1=df_offence1.astype(str)
offloc=pd.merge(df_offence1, df_allloc, how='left', left_on='Location_Code',right_on='Location_Code')
offloc[['Offence_Month','Camera_Type_x','Location_Code','Offence_Desc','Location_Desc','LATITUDE','LONGITUDE']]




```

```python
offloc['Offence_Month']=pd.to_datetime(offloc['Offence_Month'])
offloc['Year']=offloc['Offence_Month'].dt.year
```

```python

# checking the locations with no coordinates
nullgroups=set(offloc[offloc['LATITUDE'].isnull()]['Location_Code'])



checknull=[]
for each in nullgroups:
    if each not in allcode:
        checknull.append(each)

        
checkloc=offloc[offloc['Location_Code'].isin(checknull)][['Location_Code','Location_Desc','Camera_Type_x']].drop_duplicates()
        
print(checkloc) 
```

from the results above we can see the coordinates for 1003 can be identify as it is an interaction base don location description, update the latitude and longitude based on the coordinates found on the google map

```python

checkloc1=offloc[offloc['Location_Code'].isin(checknull)]

checkloc1[['Offence_Month','Camera_Type_x','Offence_Desc','Location_Code','Location_Desc','LATITUDE','LONGITUDE']]
 
alist=list(offloc[offloc['Location_Code']=='1003'].index.values) 

for i in alist:
    offloc.at[i,'LATITUDE']= ('-35.235517')
    offloc.at[i,'LONGITUDE']= ('149.058055')


offloc['LATITUDE']=offloc['LATITUDE'].astype('float')
offloc['LONGITUDE']=offloc['LONGITUDE'].astype('float')
```

# Export it as a new df

```python

offloc.to_csv ('merged_offence_location.csv', index = False, header=True)
```

# Car crash data read from XML

```python

import xml.etree.ElementTree as ET
import pandas as pd

xtree = ET.parse('car_crash.xml')
root=xtree.getroot()

item_list=['crash_id','crash_date','x','y','crash_severity']


def get_xml(root=root,item_list=item_list,length=63692):
    big_item_list=[]
    for each in item_list:
        big_item_list.append([])

    for i in range(length):
        for each in item_list:
            for x in root[0][i]:
                if x.tag==each:
                    big_item_list[item_list.index(each)].append(x.text)

    big_item=dict()
    i=0
    for i in range(len(big_item_list)):
        big_item[item_list[i]]=big_item_list[i]
    return big_item


df=pd.DataFrame(get_xml())


df.to_csv ('crash_dataframe.csv', index = False, header=True)

```
